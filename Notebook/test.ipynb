```python
import os
data=os.path.join("dataset","algerian_forest_fires.csv")
df=pd.read_csv(r"D:\PROJECT\dataset\Algerian_forest_fires_cleaned_dataset.csv")
df.head()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.drop(['day','month','year','Region'],axis=1,inplace=True)
```

```python
X=df.drop(columns=['Classes'])
y=df['Classes']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
X_train.shape, X_test.shape
```

```python
import pickle
from pathlib import Path
def scaler_standard(X_train,X_test):
    scaler= StandardScaler()
    X_train_scaled= scaler.fit_transform(X_train)
    X_test_scaled= scaler.transform(X_test)

    file= open(r'D:\PROJECT\Model\StandardScaler.pkl','wb')
    pickle.dump(scaler,file)
    file.close()

    return X_train_scaled, X_test_scaled
```

```python
X_train_scaled,X_test_scaled= scaler_standard(X_train,X_test)
```

```python
X_train_scaled
```

```python
log_reg= LogisticRegression()
log_reg.fit(X_train_scaled,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
import warnings
warnings.filterwarnings('ignore')
parameters= {
    'penalty':['l1','l2'],
    'C' : np.logspace(-3,3,7),
    'solver':['newton-cg','lbfgs','liblinear'],
}
```

```python
logreg= LogisticRegression()
clf= GridSearchCV(logreg,
                  param_grid = parameters,
                  scoring = 'accuracy',
                  cv = 10)

clf.fit(X_train_scaled,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
y_pred=clf.predict(X_test_scaled)
```

```python
conf_mat= confusion_matrix(y_test,y_pred)
conf_mat
```

```python
True_positive= conf_mat[0][0]
False_positive= conf_mat[0][1]
True_negative= conf_mat[1][1]
False_negative= conf_mat[1][0]
```

```python
accuracy= (True_positive+True_negative)/(True_positive+False_positive+False_negative+True_negative)
accuracy
```

```python
precision= True_positive/(True_positive+False_positive)
precision
```

```python
recall= True_positive/(True_positive+False_negative)
recall
```

```python
F1_score= 2*(recall*precision)/(recall+precision)
F1_score
```
